### 1. Load dataset from csv file - 1 point

Note: Please load the dataset from the URL as shown in the demo 

```python
# Import needed libraries to get data from URL and load into Pandas DF
import os
import tarfile
import urllib
import pandas as pd

# Define url and OS path that I want to save csv data under
DATA_URL = "https://raw.githubusercontent.com/JacobBRodriguez/SJSU/master/Fall%202020/DataSets/datasetHW2.csv"
DATA_PATH = os.path.join("/Users/Do-While/Desktop/SJSU/Classes/SJSU/Fall 2020", "DataSets")

# Function to get data from URL and return it as Pandas dataframe
def fetch_and_load_from_url(data_url, data_path):
    
    # If no data path exists, create directory
    if not os.path.isdir(DATA_PATH):
        os.makedirs(DATA_PATH)
    # create text document path under OS path
    txt_path = os.path.join(DATA_PATH, "hw2_data.txt")
    # Using URL, request data and return to created path
    urllib.request.urlretrieve(DATA_URL, txt_path)
    # Read in text file, convert csv to pandas df and return upon close of file
    with open(txt_path, "r") as file:
        return pd.read_csv(file, header=None)
    
hwDataSet = fetch_and_load_from_url(DATA_URL, DATA_PATH)
print(hwDataSet)
```

### 8. Perform min-max normalization on features and target - 2 points

```python
# Will be dropping the last column, since it is a string column and doesn't make sense to normalize.
dropped_last = new_df.drop(columns=9)

# Need to also convert columns to numeric
# Selecting all columns
cols = dropped_last.columns
# Applying to_numeric function to all selected columns of data
dropped_last[cols] = dropped_last[cols].apply(pd.to_numeric)

# Using a 0 - 1 normalization with the standard function of Xnew = X - Xmin / Xmax - Xmin
normalized_df=(dropped_last-dropped_last.min())/(dropped_last.max()-dropped_last.min())
normalized_df
```

### 9. Describe rules of NumPy Broadcasting - 1 point

```python
# When dealing with two arrays (or matrices), the trailing dimensions must either:
# 1) Be equal to eachother (e.g. 5 x 4 * 4) OR
# 2) One of them must be one (e.g. 5 x 4 * 5 x 1)

# This will allow for arithmetic operations to be performed on matrices of different dimensions.
# In a sense, the smaller dimensioned array can be stretched or broadcasted to fit the dimensions of the larger matrix
# without the need for a fully filled out matrix. Numpy does not create additional memory to accomplish this so the operation
# is typically faster and uses less memory than if you were to do arithmetic with two matrices of the same size, with one
# being nothing but repeating scalar values.
```

### 10. Describe the purpose of numpy.reshape with a valid example - 0.5 points


```python
# reshape will take an array or matrix and change it to a newly requested dimensional array without changing the data

# An example could be a 2 x 3 matrix but you want it as 3 x 2 so as to perform matrix multiplication to another matrix
x = np.array([[2,3,4], [5,6,7]])
print("Original matrix\n",x)
x_new = np.reshape(x, (3,2))
print("Reshaped matrix\n",x_new)
```
