```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
# Creating DTree classifier object and checking parameters for object
clf = DecisionTreeClassifier()
clf.get_params()
```

```python
# Loading the Digits dataset
digits = datasets.load_digits()
```

```python
param_grid = {'splitter': ['best', 'random'], 'max_depth': [1,2,3,4,5,6,7,8,9],'criterion': ['gini', 'entropy']}

clf_grid = GridSearchCV(estimator=clf, param_grid=param_grid)
```

```python
import time
start_time_grid = time.time()
clf_grid.fit(digits.data, digits.target)
end_time_grid = time.time()
print("Best parameter settings: ",clf_grid.best_params_)
```

### Experiment with optimizing precision, recall and f1

```python
# First precision
scoring = {'precision': sklearn.metrics.make_scorer(sklearn.metrics.precision_score, average = 'macro'),
           'recall': sklearn.metrics.make_scorer(sklearn.metrics.recall_score, average = 'macro'),
           'f1_macro': sklearn.metrics.make_scorer(sklearn.metrics.f1_score, average = 'macro')}

clf_precision = GridSearchCV(estimator=clf,param_grid=param_grid, scoring=scoring, refit="precision")
clf_precision.fit(digits.data, digits.target)
print("Best parameter settings: ",clf_precision.best_params_)
```

```python
clf_precision.best_score_
```

```python
clf_precision.cv_results_
```

```python
clf_recall = GridSearchCV(estimator=clf,param_grid=param_grid, scoring=scoring, refit="recall")
clf_recall.fit(digits.data, digits.target)
print("Best parameter settings: ",clf_recall.best_params_)

clf_recall.best_score_
```

```python
clf_f1 = GridSearchCV(estimator=clf,param_grid=param_grid, scoring=scoring, refit="f1_macro")
clf_f1.fit(digits.data, digits.target)
print("Best parameter settings: ",clf_f1.best_params_)

clf_f1.best_score_
```

## Same as above but with RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

clf_random = RandomizedSearchCV(estimator=clf, param_distributions=param_grid)
start_time_random = time.time()
clf_random.fit(digits.data, digits.target)
end_time_random = time.time()
print("Best parameter settings: ",clf_random.best_params_)
```

```python
# First precision
scoring = {'precision': sklearn.metrics.make_scorer(sklearn.metrics.precision_score, average = 'macro'),
           'recall': sklearn.metrics.make_scorer(sklearn.metrics.recall_score, average = 'macro'),
           'f1_macro': sklearn.metrics.make_scorer(sklearn.metrics.f1_score, average = 'macro')}

clf_precision = RandomizedSearchCV(estimator=clf,param_distributions=param_grid, scoring=scoring, refit="precision")
clf_precision.fit(digits.data, digits.target)
print("Best parameter settings: ",clf_precision.best_params_)
```

```python
clf_precision.best_score_
```

```python
# Next Recall
clf_recall = RandomizedSearchCV(estimator=clf,param_distributions=param_grid, scoring=scoring, refit="recall")
clf_recall.fit(digits.data, digits.target)
print("Best parameter settings: ",clf_recall.best_params_)

clf_recall.best_score_
```

```python
# Lastly F1
clf_f1 = RandomizedSearchCV(estimator=clf,param_distributions=param_grid, scoring=scoring, refit="f1_macro")
clf_f1.fit(digits.data, digits.target)
print("Best parameter settings: ",clf_f1.best_params_)

clf_f1.best_score_
```

```python
print("Time taken for Grid Search: ",end_time_grid - start_time_grid)
print("Time taken for Randomized Search: ", end_time_random - start_time_random)
```
