```python
from orion.data import load_signal
from orion import Orion

train_data = load_signal('S-1-train')
train_data.head()
```

```python
from orion import Orion

hyperparameters = {
    'keras.Sequential.LSTMTimeSeriesRegressor#1': {
        'epochs': 5,
        'verbose': True
    }
}

orion = Orion(
    pipeline='lstm_dynamic_threshold',
    hyperparameters=hyperparameters
)

orion.fit(train_data)
```

```python
new_data = load_signal('S-1-new')
anomalies = orion.detect(new_data)
```

## New dataset

```python
dataset_aqi = pd.read_csv('./CMPE-258-Anomaly-Detection/daily_aqi_by_cbsa_2020.csv', delimiter=",")
```

### Convert column to datetime object

```python
dataset_aqi['Date_new'] = pd.to_datetime(dataset_aqi['Date'], format=('%Y/%m/%d'))
```

```python
dataset_aqi
```

### Convert to Epoch

```python
dataset_aqi['Epoch'] = (dataset_aqi['Date_new'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')
```

### Format and rename

```python
dataset_aqi = dataset_aqi.fillna(0)
dataset_aqi['Epoch'].astype(np.int64)


dataset_aqi['timestamp'] = dataset_aqi['Epoch'].astype(np.int64)

final_out = dataset_aqi[['timestamp', 'AQI']].copy()

final_out = final_out.rename(columns={"AQI":"value"})
```

### Get output

```python
new_anomals = orion.detect(final_out)
new_anomals
```
